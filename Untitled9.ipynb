```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the data into a dataframe
df = pd.read_csv("Growlocations.csv")
```

```python
# Filter the DataFrame to only keep rows where the 'Latitude' is between -10.592 and 1.6848 and the 'Longitude' is between 50.681 and 57.985

df = df[(df['Latitude'] >= -10.592) & (df['Latitude'] <= 1.6848) & (df['Longitude'] >= 50.681) & (df['Longitude'] <= 57.985)]
```

```python
# Load the map image
im = plt.imread("map.png")
```

```python
# Create a new figure and subplot
figure, ax = plt.subplots()
# Plot a scatter plot of the 'Latitude' and 'Longitude' columns on the subplot
ax.scatter(df['Latitude'], df['Longitude'])
# Read in the map image and store it in a variable
map2 = plt.imread('map.png')
# Display the image on the subplot, setting the extent of the image to the specified coordinate
ax.imshow(map2, extent=(-10,1.7,50,57))
# Save the figure to an image file
figure.savefig('Mapcom.png')
```
